### Đọc dữ liệu

```python
df=pd.read_csv('Europe_new.csv')
df.head(10)
```

### Tỉ lệ sản xuất các loại điện năng ở châu Âu 

#### Xử lý dữ liệu

##### Bước 1
+ Lấy danh sách các loại điện có trong data frame
+ Lấy danh sách các loại điện sản xuất ở cột commodity_transaction	
+ lọc ra các hàng có liên quan đến dữ liệu điện

```python
electric=['hydro','nuclear_electricity','solar_electricity','thermal_electricity',
        'tide_wave_and_ocean_electricity','total_electricity','wind_electricity']

electric_df=df[df['category'].isin(electric)]
```

##### Bước 2
+ lọc ra data frame chỉ có sản xuất điện
+ lọc ra data frame chỉ có sử dụng điện

```python
consumption_df=electric_df[electric_df['commodity_transaction']=='Electricity - Gross production']
production_df=electric_df[electric_df['commodity_transaction']=='Electricity - Gross demand']
```

##### Bước 3
+ gom nhóm các loại điện của các quốc gia lại với nhau
+ chia cho tổng để có tỉ lệ phân trăm

```python
rate_production=electric_df.groupby('category')['quantity'].sum()
rate_production.drop(index='total_electricity',inplace=True)
rate_production=rate_production*100/rate_production.sum()
rate_production
```

##### Bước 4
+ Vì " tide_wave_and_ocean_electricity " quá nhỏ nên không thể hiện được cho nên chúng ta tạm thời xóa nó ra khỏi bảng
+ Vẽ pie char để thể hiện tỉ lệ của các loại điện

```python
rate_production.drop(index='tide_wave_and_ocean_electricity',inplace=True)
```

```python
palette_color=sns.color_palette('bright')
fig, ax=plt.subplots(figsize=(14,8))
ax.pie(rate_production[0:5].values,labels=rate_production[0:5].index,colors=palette_color)
ax.set_title("Tỉ lệ sản xuất các loại điện ở châu Âu")
plt.legend(loc=4,fontsize=6, title_fontsize=20);
```

Nhìn vào biểu đồ có thể thấy nhiệt điện và điện hạt nhân là được sản xuất nhiều nhất

### Các quốc gia nào sử dụng điện nhiều nhất

##### Bước 1:
+ chúng ta có " consumption_df " là data frame sử dụng điện của các quốc gia
+ Gom nhóm các loại điện của 1 quốc gia lại
+ Lấy 10 quốc gia sử dụng điện nhiều nhất

```python
country_consumption=consumption_df.groupby('country_or_area')['quantity'].sum()
T=country_consumption.sort_values(ascending=False)
top_10=T[:10]
top_10
```

##### Bước 2
+ Vẽ biểu đồ

```python
fig, ax=plt.subplots(figsize=(14,5))
ax.barh(top_10.index,top_10.values)
ax.set_title('Top 10 quốc gia sử dụng điện nhiều nhất châu Âu')
ax.set_xlim([0,30000000])
ax.set_xlabel('Kwh')
for x,y in zip(top_10.values,top_10.index):
    plt.text(x+100000, y, x)
```

Nhìn vào biểu đồ có thể thấy top 10 quốc gia sử dụng điện nhiều nhất là:
1. Russian Federation
2. Germany
3. France
4. United Kingdom
5. Italy
6. Spain
7. Sweden
8. Poland
9. Netherlands
10. Belgium

##### Top 5 quốc gia xuất khẩu điện và top 5 quốc gia nhập khẩu điện

##### Bước 1
+ Lấy các hàng có giá trị là " Electricity - exports " trong cột commodity_transaction
+ Gom nhóm các hàng có cùng quốc gia lại và cộng tổng giá trị của cột quantity
+ Lấy các hàng có giá trị là " Electricity - imports " trong cột commodity_transaction
+ Gom nhóm các hàng có cùng quốc gia lại và cộng tổng giá trị của cột quantity
+ Lấy xuất khẩu trừ nhập khẩu rồi sắp xếp giảm dần
+ Lấy 5 giá trị đầu tiên là top 5 xuất khẩu
+ Lấy 5 giá trị cuối cùng là nhập khẩu
+ Nối 2 series lại vs nhau và sắp xếp lại theo giá trị tăng dần

```python
exports=electric_df[electric_df['commodity_transaction']=='Electricity - exports']
exports=exports.groupby('country_or_area')['quantity'].sum()
imports=electric_df[electric_df['commodity_transaction']=='Electricity - imports']
imports=imports.groupby('country_or_area')['quantity'].sum()
difference_values=(exports-imports).sort_values(ascending=False)
top_5_exports=difference_values[0:5]
top_5_imports=difference_values[-5:]
T=pd.concat([top_5_exports,top_5_imports],axis=0)
T=T.sort_values(ascending=True)
```

##### Bước 2 
+ vẽ hình

```python
fig, ax=plt.subplots(figsize=(14,5))
ax.barh(T.index,T.values)
ax.set_xlim(-1500000,2000000)
for x,y in zip(T.index,T.values):
    if y<0:
        plt.text(y-300000,x,y)
    else:
        plt.text(y+50000,x,y)
ax.set_title("Top 5 quốc gia xuất khẩu điện và nhập khẩu điện")
ax.set_xlabel('kwh');
```

Nhìn vào biểu đồ có thể thấy 

Top 5 quốc gia xuất khẩu điện là:
1. France
2. Nga
3. Czechia
4. Germany
5. Poland

Top 5 quốc gia nhập khẩu điện
1. Hungary
2. Finland
3. UK
